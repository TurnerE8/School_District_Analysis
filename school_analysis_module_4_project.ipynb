```python
#DELIVERABLE 1
#Read a CSV File 

##Import the libraries we will need to use in order to run our code 
import pandas as pd 
import os 

##Find the CSV file and then read the CSV file 
student_data = os.path.join('..', 'Resources', 'new_full_student_data.csv')

student_df = pd.read_csv(student_data)

###we want to make sure Pandas read in our data correctly by using the 'head' function below 
student_df.head()
```

```python
#DELIVERABLE 2
###check the data for any NaN values 
student_df.isna().sum()

###drop any rows that contain NaN values and verify that no NaN values remain by running the following code: 
student_df = student_df.dropna()

student_df.isna().sum()

###check data for any duplicated rows in the data 
student_df.duplicated().sum()

###drop all of the duplicated rows and check to make sure the data has dropped the numbered of duplicated rows 
student_df = student_df.drop_duplicates()

student_df.duplicated().sum()

###check the data types of the columns, specifically the "grade" column 
student_df.dtypes

###visually inspect the "grade" column and observe that its datatype is an object rather than an integer 
student_df['grade']

###change the "grade" column to drop the 'th' at the end of the grade number by using the str.replace function 
student_df['grade'] = student_df['grade'].str.replace('th', '')

student_df['grade']

###change the data type of the "grade" column to int, rather than object using the astype function 
student_df['grade'] = student_df['grade'].astype(int)
student_df.dtypes
```

```python
#DELIVERABLE 3
###Generate a General Summary Statistic for the student data
student_df.describe()
```
